```python
from TradingviewData import TradingViewData, Interval
import pandas as pd
import os
```

```python
tv = TradingViewData()
```

```python
xrp_data = tv.get_hist(
    symbol='XRPUSD',       
    exchange='BITSTAMP',  
    interval=Interval.daily,
    n_bars=1800            
)
```

El "problema" del método get_hist() es que tenemos que indicar en n_bars el nº de datos que queremos de antemano. Elegimos más de la cuenta para luego hacer filtros por años 

```python
years = [2022, 2023, 2024, 2025]

for year in years:
    for month in range(1, 13):
        month_path = f'data/{year}/{month:02d}'
        os.makedirs(month_path, exist_ok=True)

        mask = (xrp_data.index.year == year) & (xrp_data.index.month == month)
        df_month = xrp_data.loc[mask]
        
        if not df_month.empty:
            ruta = f'{month_path}/XRP_{year}_{month:02d}.csv'
            df_month.to_csv(ruta, index=True)
            # print(f'Archivo guardado: {ruta} ({len(df_month)} datos)')
```
